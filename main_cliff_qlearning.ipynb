## Import + Setup


```python
import cliff_qlearning_agent
import gymnasium as gym
import matplotlib.pyplot as plt
import pandas as pd
import time
from IPython.display import display, clear_output
```

## Environment + Agent
Laden der virtuellen Umgebung und initialisieren der Parameter.

```python
# env = gym.make("CliffWalking-v1", render_mode="human")
env = gym.make("CliffWalking-v1", render_mode="rgb_array")

agent = cliff_qlearning_agent.CliffWalkingAgent(
    env=env,
    learning_rate=0.1,
    initial_epsilon=1.0,
    epsilon_decay=1.0 / (5000 / 2),
    final_epsilon=0.1,
    discount_factor=0.95,
    visualize_training=False, 
)
```

## Training
Durch diese Funktion wird der Agent in der virtuellen Umgebung trainiert.

```python
agent.train_q_learning(episodes=3000)
```

## Demo
Hier werden die einzelnen Bewegungsframes gesammelt, um später eine Visualisierung
durchführen zu können.

```python
demo_frames = agent.run_policy_demo()
```

## Visualisierung
Nun können die Bewegungen des Agents visualisiert werden.

```python
for f in demo_frames:
    plt.imshow(f)
    plt.axis("off")
    display(plt.gcf())
    clear_output(wait=True)
    time.sleep(0.1)
```
